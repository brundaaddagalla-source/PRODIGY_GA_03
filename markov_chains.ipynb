```python
import wikipedia, markovify
import ipywidgets as w
from IPython.display import display, Markdown
```

```python
topics=["Artificial intelligence", "Machine learning", "Natural Language processing", "Data Science", "Information retrieval"]
text=""
for i in topics:
  try:
    text+=wikipedia.page(i).content
  except:
    pass
print("Dataset size (characters): ",len(text))
```

```python
model=markovify.Text(text, state_size=2)
```

```python
seed_input=w.Text(description="Start word:", placeholder="Enter a word: ", style={'description_width':'initial'})
sentence=w.IntSlider(value=3, min=1,max=10,description="Text generated: ",style={'description_width':'initial'})
generate=w.Button(description='Generate Text', button_style='success')
o=w.Output()
```

```python
def generate_text(b):
  o.clear_output()
  sw = seed_input.value.strip()

  with o:
    s=[]
    for i in range(sentence.value):
      if sw:
        j=model.make_sentence_with_start(sw, strict=False)
      else:
        j=model.make_sentence()
      if j:
        s.append(j)
    display(Markdown("### Generated Text"))
    print(" ".join(s))
generate.on_click(generate_text)
```
